En este caso escogí la BD de Casos COVID en Colombia que está publicada en https://www.datos.gov.co/Salud-y-Protecci-n-Social/Estado-de-Casos-de-Coronavirus-COVID-19-en-Colombi/6c4c-msrp


```
import csv
import json
```

```
with open('Casos_COVID_Colombia.csv', 'r', encoding='UTF-8') as covid, open('Analisis_COVID.csv', 'w', encoding='UTF-8') as analisis:
  lector = csv.DictReader(covid)
  campos = lector.fieldnames
  campos[0] = 'ID de caso'
  next(lector, None)

  diccionario = {}
  lista = []
  lista_actual = []
  region_actual = ''

  for caso in lector:
    depart = caso["Departamento o Distrito "]
    sex = caso["Sexo"]
    if region_actual != depart:
      diccionario["Regiones"] = lista_actual
      diccionario = {
          "Region":depart
      }
      lista.append(diccionario)
      lista_actual = []
    lista_actual.append({"Sexo":sex})
    region_actual = depart
  
  print(json.dumps(lista, indent=4, ensure_ascii=False))
```
