```
#Ahora vamos a calcular cuanto fue el valor de cada una de las asociaciones

with open('subvenciones - Hoja1.csv', encoding='latin1') as fichero:
  lector = csv.reader(fichero)
  next(lector, None)
  diccionario = {
      "JUANES":2400
  }

  for linea in lector:
    asociacion = linea[0]
    importe = float(linea[2])
    if asociacion in diccionario: #valida si algo está contenido en un objeto
      diccionario[asociacion] = diccionario[asociacion] + importe #valor actual + importe
    else:
      diccionario[asociacion] = importe
  

  print(json.dumps(diccionario, indent=4, sort_keys=True))
```

```
#con DictReader voy a leer los datos directamente como un diccionario
# y me toma la primera línea como encabezado

with open('subvenciones - Hoja1.csv', encoding='latin1') as fichero:
  lector = csv.DictReader(fichero)
  next(lector, None)
  diccionario = {
      "JUANES":2400
  }

  for linea in lector:
    asociacion = linea["AsociaciÃ³n"]
    importe = float(linea["Importe"])
    if asociacion in diccionario: #valida si algo está contenido en un objeto
      diccionario[asociacion] = diccionario[asociacion] + importe #valor actual + importe
    else:
      diccionario[asociacion] = importe
  
  print(json.dumps(diccionario, indent=4, sort_keys=True))
```

```
#Editando las líneas del archivo con DictReader y DictWriter

with open('subvenciones - Hoja1.csv', encoding='latin1') as fichero_csv, open('subvenciones_just.csv', 'w', encoding='latin1') as fichero_esc:
  lector = csv.DictReader(fichero_csv)
  campos = lector.fieldnames + ['Justificacion requerida', 'Justificacion recibida']
  escritor = csv.DictWriter(fichero_esc, fieldnames= campos)
  escritor.writeheader()
  for linea in lector:
    importe = float(linea['Importe'])
    if importe > 350:
      linea['Justificacion requerida'] = 'Si'
    else:
      linea['Justificacion requerida'] = 'No'
    linea['Justificacion recibida'] = 'No'
    escritor.writerow(linea)
```

### TSV
Los TSV son lo mismo que los CSV pero que se separan por Tabulaciones. Es muy útil porque los valores pueden tener comas (,) en el interior.

```
#La librería csv también permite trabajar con tsv

with open('subvenciones - subvenciones.tsv', encoding='UTF-8') as miFicherito:
  lector = csv.DictReader(miFicherito, delimiter = '\t')
  next(lector, None)
  diccionario = {
      "JUANES":2400
  }

  for linea in lector:
    asociacion = linea["Asociación"]
    importe = float(linea["Importe"])
    if asociacion in diccionario: #valida si algo está contenido en un objeto
      diccionario[asociacion] = diccionario[asociacion] + importe #valor actual + importe
    else:
      diccionario[asociacion] = importe
  
  print(json.dumps(diccionario, indent=4, sort_keys=True))
```

Vamos a convertir este json en un archivo más legible

```
with open('subvenciones.json', encoding='UTF-8') as fichero_json, open('subvenciones_agrupadas.json', 'w', encoding='UTF-8') as fichero_esc:
  datos=json.load(fichero_json)
  lista = []
  lista_actual = []
  asociacion_actual = ''
  diccionario = {}
  for elemento in datos: #ciclo por cada uno de los registros
    asoc = elemento["Asociación"]
    imp = elemento["Importe en euros"]
    act = elemento["Actividad Subvencionada"]
    if(asociacion_actual != asoc):
      diccionario["Actividades"] = lista_actual
      diccionario = {
          "Asociación":asoc
      }
      lista.append(diccionario)
      lista_actual = []
    lista_actual.append({"Actividad Subvencionada":act, "Importe en euros":imp})
    asociacion_actual = asoc
  
  print(json.dump(lista, fichero_esc, indent=4, ensure_ascii=False))

  #ensure_ascii hace que las tildes se vean correctamente
```

### Excel

Python no tiene una libreria estandar para excel por lo que se usará las bibliotecas xlrd y xlwt para lectura y escritura respectivamente. Estas librerías pueden ser consideradas estandares de facto para estas operaciones sobre archivos, Pandas se puede utilizar para leer archivos en excel.

```
from xlrd import open_workbook, colname
import xlwt
```

```
with open_workbook('subvenciones.xlsx', on_demand=True) as libro:
  asociaciones = {}
  for nombre in libro.sheet_names():
    hoja = libro.sheet_by_name(nombre)
    for numeroDeFila in range(1, hoja.nrows):
      fila = hoja.row(numeroDeFila)
      asociacion = fila[0].value
      importe = fila[2].value
      if asociacion in asociaciones:
        asociaciones[asociacion]=asociaciones[asociacion]+importe
      else:
        asociaciones[asociacion] = importe
  
  print(asociaciones)
```

```
import pandas as pd
```

```
with pd.ExcelFile('subvenciones.xlsx') as xl:
  writer = pd.ExcelWriter('subvenciones_pd.xlsx')
  for nombre in xl.sheet_names:
    df = xl.parse(nombre)
    df.to_excel(writer, nombre)
  writer.save
```

### XML

Se uba biblioteca etree

```
import xml.etree.ElementTree as ET
```

```
arbol = ET.parse('subvenciones.xml')
raiz = arbol.getroot()
asociaciones = {}

for fila in raiz:
  asociacion = fila[0].text
  importe = float(fila[2].text)
  if asociacion in asociaciones:
    asociaciones[asociacion] = asociaciones[asociacion] + importe
  else:
    asociaciones[asociacion] = importe
  
print(asociaciones)
```
