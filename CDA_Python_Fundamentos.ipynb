```
def cuadrado(x):
  lista = []
  for e in x:
    lista.append(e ** 2)
    print('-->', lista)
  print(lista) #esto no es eficiente
  
cuadrado([1,2,3,4])
```

```
def square(x):
  return [e ** 2 for e in x] #esta forma si es mucho más eficiente

square([1,2,3,4])
```

```
def mayores(x):
  result = []
  for e in x:
    if e >5:
      result.append(e)
  return result #no es eficiente

mayores([1,2,3,4,5,6,7])
```

```
def mayores(x):
  return [e for e in x if e > 5] #forma eficiente

mayores([1,2,3,4,5,6,7])
```

```
# definir una lista de listas como una sola lista
def matriz(x):
  result = []
  for row in x:
    for e in row:
      result.append(e)
  return result      #forma no eficiente

matriz(x)
```

```
def matrix(x):
  return [e for row in x for e in row] #forma eficiente

matrix(x)
```
